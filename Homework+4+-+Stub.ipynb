```python
import matplotlib
import matplotlib.pyplot as plt
import nltk

# Slice dataframe
US=U[U['country']=='us']

# Clean data
txt = US.comments.str.lower().str.replace(r'\|', ' ').str.cat(sep=' ')

# Get words from data
words = nltk.tokenize.word_tokenize(txt)
word_dist = nltk.FreqDist(words)

# Get stopwords from words
stopwords = nltk.corpus.stopwords.words('english')
other_words = nltk.FreqDist(w for w in words if w not in stopwords) 

print('All frequencies, including STOPWORDS:')

top = 10
rslt = pd.DataFrame(word_dist.most_common(top),
                    columns=['Word', 'Frequency'])
print(rslt)

rslt = pd.DataFrame(other_words.most_common(top),
                    columns=['Word', 'Frequency']).set_index('Word')

matplotlib.style.use('ggplot')

rslt.plot.bar(rot=0)
```
